```python
import pandas as pd
import matplotlib.pyplot as plt
import descartes
import geopandas as gpd
from shapely.geometry import Point, Polygon

%matplotlib inline
```

```python
world_map = gpd.read_file('world.shp')
```

```python
df = pd.read_csv('COVID-19.csv')
crs = {'init': 'epsg:4326'}
```

```python
geometry = [Point(xy) for xy in zip(df['Long'], df['Lat'])]
```

```python
geo_df = gpd.GeoDataFrame(df,
                         crs = crs,
                         geometry = geometry)
```

```python
fig, ax = plt.subplots(figsize = (20,20))
world_map.plot(ax = ax, alpha = 0.4, color = 'skyblue')
i = 4
geo_df[geo_df['Confirmed']<50].plot(ax=ax, markersize=i, c = 'green', marker = "o", label='less than 50 cases')
geo_df[geo_df['Confirmed']>50].plot(ax=ax, markersize=i, color = 'steelblue', marker = "o", label='more than 50 cases')
geo_df[geo_df['Confirmed']>100].plot(ax=ax, markersize=i, color = 'blue', marker = "o", label='more than 100 case')
geo_df[geo_df['Confirmed']>150].plot(ax=ax, markersize=i, color = 'yellow', marker = "o", label='more than 150 cases')
geo_df[geo_df['Confirmed']>200].plot(ax=ax, markersize=i, color = 'purple', marker = "o", label='more than 200 cases')
geo_df[geo_df['Confirmed']>250].plot(ax=ax, markersize=i, color = 'red', marker = "o", label='more than 250 cases')
geo_df[geo_df['Confirmed']>300].plot(ax=ax, markersize=i, color = 'pink', marker = "o", label='more than 300 cases')
geo_df[geo_df['Confirmed']>350].plot(ax=ax, markersize=i, color = 'brown', marker = "o", label= 'more than 350 cases')
geo_df[geo_df['Confirmed']>400].plot(ax=ax, markersize=i, color = 'black', marker = "o", label= 'more than 400 cases')
geo_df[geo_df['Confirmed']>450].plot(ax=ax, markersize=i, color = 'orange', marker = "o", label= 'more than 450 cases')





plt.legend(prop={'size':15})
```
